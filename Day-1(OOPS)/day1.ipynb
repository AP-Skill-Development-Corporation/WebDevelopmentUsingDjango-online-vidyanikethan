```python
### parametrized constructor:

class Math:
    def __init__(self,a,b):
        self.v1=a
        self.v2=b       ##storage
    def show(self):
        print(self.v1)   ### display the values
        print(self.v2)
    def add(self):            ### addition
        return self.v1+self.v2
    
obj=Math(4,7)
print(obj.show())
print(obj.add())
```

```python
####single level inheritance

class ClassA():
    a,b=20,30
    def sample():
        print("im from ClassA")
class ClassB(ClassA):
    c,d=40,50
    def sample2():
        print("im from ClassB")
        
obj=ClassB
print(obj.a)
print(obj.b)
print(obj.sample())
print(obj.sample2())
```

```python
### multilevel:

class A:
    def classA():
        print("im from class A")
        
class B(A):
    def classB():
        print("im from class B")
class C(B):
    def classC():
        print("im from class C")

obj=C
print(obj.classA())
print(obj.classB())
print(obj.classC())
        
        
        
```

```python
### Multiple
class A:
    def classA():
        print("im from class A")
        
class B:
    def classB():
        print("im from class B")
class C(A,B):
    def classC():
        print("im from class C")

obj=C
print(obj.classA())
print(obj.classB())
print(obj.classC())
        
```
